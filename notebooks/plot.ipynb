```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 图片路径
img_paths = [
    "../figures/guests_all.png",
    "../figures/guests_adults_filter.png",
    "../figures/guests_adults_and_children_filter.png"
]

labels = ['A', 'B', 'C']  # 小图标号

# 创建三行一列的子图
fig, axes = plt.subplots(3, 1, figsize=(18, 10))  # 3行1列

for ax, path, label in zip(axes, img_paths, labels):
    img = mpimg.imread(path)
    ax.imshow(img)
    ax.axis('off')
    ax.text(-0.02, 0.95, label, transform=ax.transAxes, fontsize=10, fontweight='bold',
            color='black', va='top', ha='left', bbox=dict(facecolor='white', alpha=0.5, edgecolor='none'))

plt.tight_layout()
plt.savefig("../figures/guests_comparison.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.gridspec import GridSpec

# 图片路径
img_paths = [
    "../figures/nights_distribution_99p.png",
    "../figures/nights_sf_and_cancellation.png",
    "../figures/nights_cancellation_week_end.png"
]

labels = ['A', 'B', 'C']  # 小图标号

# 创建 Figure
fig = plt.figure(figsize=(18, 12))  # 总画布大小

# 使用 GridSpec，height_ratios 控制每个子图高度比例
gs = GridSpec(nrows=3, ncols=1, height_ratios=[1, 1.2, 1.8], figure=fig)

# 循环绘图
for i, (path, label) in enumerate(zip(img_paths, labels)):
    ax = fig.add_subplot(gs[i, 0])
    img = mpimg.imread(path)
    ax.imshow(img)
    ax.axis('off')
    
    # 小图标号
    ax.text(-0.02, 0.95, label, transform=ax.transAxes, fontsize=10, fontweight='bold',
            color='black', va='top', ha='left', bbox=dict(facecolor='white', alpha=0.5, edgecolor='none'))

plt.tight_layout()
plt.savefig("../figures/nights_comparison.png", dpi=300, bbox_inches='tight')
plt.show()
```
